## Munge Summary Statistics

```python
from ldsc import MungeConfig, RawSumstatsSpec, SumstatsMunger
```

```python
# Set these explicitly for your environment.
DATA_DIR = Path("/absolute/path/to/resources")
OUTPUT_DIR = Path("/absolute/path/to/result_consistency_checks/output_new")

SUMSTATS_PATH = DATA_DIR / "sumstats_raw" / "pgc-mdd2025_no23andMe-noUKBB_eur_v3-49-24-11_clean.tsv"
OUT_PREFIX = OUTPUT_DIR / "sumstats_processed" / "mdd2025"

for path_name, path_value in {
    "DATA_DIR": DATA_DIR,
    "OUTPUT_DIR": OUTPUT_DIR,
    "SUMSTATS_PATH": SUMSTATS_PATH,
}.items():
    if not path_value.exists():
        raise FileNotFoundError(f"Set {path_name} to an existing path before running the notebook: {path_value}")
```

```python
sumstats = SumstatsMunger().run(
    RawSumstatsSpec(
        path=SUMSTATS_PATH,
        trait_name="mdd2025",
    ),
    MungeConfig(
        out_prefix=OUT_PREFIX,
    ),
)

sumstats.summary()
```
